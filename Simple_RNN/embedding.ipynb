```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size = 10000
```

```python
sent
```

```python
## One-hot encode the sentences
one_hot_sent = [one_hot(words, voc_size) for words in sent]
```

```python
one_hot_sent
```

```python
## Word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_sent, padding='pre', maxlen=sent_length)
embedded_docs
```

```python
## Feature Representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size, dim, input_shape=(sent_length,)))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
embedded_docs
```

```python
model.predict(embedded_docs)
```
